```python
#https://docs.scipy.org/doc/numpy/reference/generated/numpy.count_nonzero.html
#Create a numerical variable from 2 genders as per the standard ratio ratio of females to males with CH in the study 2:1
#Sort genders 
#'M' = Male, 'F' = Female
gender_ch = ['F', 'M']



#Finding the probability of these in a random sample
#In the study the gender percentages are as follows: . 
gen = np.random.choice(gender_ch, 25, p=[0.67, 0.33])

# https://
print("F: ", np.count_nonzero(gen == 'F'))
print("M: ", np.count_nonzero(gen == 'M'))
```

```python
#https://docs.scipy.org/doc/numpy/reference/generated/numpy.count_nonzero.html
#Create a numerical variable from all three etiologies as they are categorical in nature
#Sort etiology of CH by birth causation 
#'E' = Ectopia, 'D' = Dyshormonogenesis and 'A' = Athyreosis
etiology_ch = ['E', 'D', 'A']



# Finding the probability of these in a random sample
# In the study the etiology percentages are as follows: Ectopia 45%, Dyshormonogenesis 24% and Athyreosis 31%. 
eti = np.random.choice(etiology_ch, 25, p=[0.45, 0.24, 0.31])

# https://
print("E: ", np.count_nonzero(eti == 'E'))
print("D: ", np.count_nonzero(eti == 'D'))
print("A: ", np.count_nonzero(eti == 'A'))
```
